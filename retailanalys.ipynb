🔹 Step 2: Data Preparation

```python
import pandas as pd
from io import StringIO

df = pd.read_csv('zara_clean_data.csv')

# Cleaning
df['price'] = pd.to_numeric(df['price'])
df['Sales Volume'] = pd.to_numeric(df['Sales Volume'])
df['scraped_at'] = pd.to_datetime(df['scraped_at'])

# Feature Engineering
df['Revenue'] = df['price'] * df['Sales Volume']
```

🔹 Step 3: Exploratory Data Analysis (EDA)
📊 Task 1: Basic Overview

```python
print(df.describe())
print(df['Product Position'].value_counts())
```

📊 Task 2: Promotion Effect
Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Promotion', y='Sales Volume', data=df)
plt.show()
```

📊 Task 3: Product Position Impact

```python
sns.barplot(x='Product Position', y='Sales Volume', data=df)
plt.xticks(rotation=30)
plt.show()
```

📊 Task 4: Price vs Sales

```python
sns.scatterplot(x='price', y='Sales Volume', hue='Promotion', data=df)
plt.show()
```

📊 Task 5: Revenue Analysis

```python
top_rev = df.nlargest(10, 'Revenue')

sns.barplot(x='Revenue', y='name', data=top_rev)
plt.show()
```

🔹 Step 4: Interactive Dashboard (Plotly)

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="price",
    y="Sales Volume",
    size="Revenue",
    color="Product Position",
    hover_data=["name"]
)

fig.show()
```
